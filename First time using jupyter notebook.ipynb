```python
import numpy as np 
import matplotlib
import matplotlib.pylab as plt 
```

```python
#Data for plotting

t = np.arange(0.0,2.0,0.01)
s = 1+ np.sin(2 * np.pi * t)

fig, ax = plt.subplots()
ax.plot(t,s)

ax.set(xlabel='time (s)',ylabel='voltage (m)',
       title='About as simple as it gets folks'
      )
ax.grid()
```
